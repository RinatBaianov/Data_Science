### 3 Error  

–ú–µ—Ç–æ–¥ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è (–ú–ú–ü) ‚Äî —ç—Ç–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥ –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏ (–≤–∫–ª—é—á–∞—è "–≤–µ—Å–∞" –≤ –º–∞—à–∏–Ω–Ω–æ–º –æ–±—É—á–µ–Ω–∏–∏), –∫–æ—Ç–æ—Ä—ã–π –Ω–∞—Ö–æ–¥–∏—Ç —Ç–∞–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –ø—Ä–∏ –∫–æ—Ç–æ—Ä—ã—Ö —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è (–∏–ª–∏ –µ—ë –ª–æ–≥–∞—Ä–∏—Ñ–º) –¥–æ—Å—Ç–∏–≥–∞–µ—Ç –º–∞–∫—Å–∏–º—É–º–∞

**–§—É–Ω–∫—Ü–∏—è –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è**

$L(Œò) = \prod\limits_{i=1}^{m}ùúé(Œò^{T}X^{i})^{y^{i}}*(1-ùúé(Œò^{T}X^{i}))^{1-y^{i}}$   -> MAX !!!       


**–õ–æ–≥–∞—Ä–∏—Ñ–º –§—É–Ω–∫—Ü–∏–∏ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è**  

$LL(Œò) = \sum\limits_{i=1}^{m}(y^{i}ln(ùúé(Œò^{T}X^{i})) + (1-y^{i})ln(1-ùúé(Œò^{T}X^{i})))$ -> MAX !!!    

**–ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –õ–æ–≥–∞—Ä–∏—Ñ–º–∞ –§—É–Ω–∫—Ü–∏–∏ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è –ø–æ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–º—É –≤–µ—Å—É**  

$\frac{‚àÇLL(Œò)}{‚àÇŒò_{j}} = \sum\limits_{i=1}^{m}(y^{i} - ùúé(Œò^{T}X^{i}))x^{i}_{j}$    
$Œò^{k+1}_{j} = Œò^{k}_{j}$ + &eta;$\sum\limits_{i=1}^{m}(y^{i} - ùúé(Œò^{T}X^{i}))x^{i}_{j}$  

**–í—ã–≤–æ–¥ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π** :    
–ï—Å–ª–∏ $p = ùúé(z)$, $z = Œò^{T}X$, —Ç–æ–≥–¥–∞  
A) $\frac{‚àÇp}{‚àÇz} = ùúé(z)(1-ùúé(z))$   
B) $ln(x)$ = $\frac{1}{x}$  
C) $\frac{‚àÇp}{Œò_{j}} = x_{j}$  



$\frac{‚àÇLL(Œò)}{‚àÇŒò_{j}} = \frac{‚àÇLL(Œò)}{‚àÇp}\frac{‚àÇp}{‚àÇz}\frac{‚àÇz}{‚àÇŒò_{j}} = (\frac{y}{p} - \frac{1-y}{1-p})‚àÇ(z)(1-‚àÇ(z))x_{j} = (\frac{y}{p} - \frac{1-y}{1-p})p(1-p)x_{j} = [y(1-p) - p(1-y)]x_{j} = (y-p)x_{j} = [y - ‚àÇ(Œò^{T}X)]x_{j}$ 

**–í—ã–≤–æ–¥ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π (–∑–∞–ø–∏—Å—ã–≤–∞–µ–º –ø–æ–¥—Ä—É–≥–æ–º—É):**

–¶–µ–ø–Ω–æ–µ –ø—Ä–∞–≤–∏–ª–æ:

$$
\frac{\partial LL(\Theta)}{\partial \Theta_j} = \sum_{i=1}^m \frac{\partial}{\partial \Theta_j} \left( y^i \ln p^i + (1 - y^i) \ln (1 - p^i) \right)
$$

–†–∞—Å–∫–ª–∞–¥—ã–≤–∞–µ–º:

$$
= \sum_{i=1}^m \left( \frac{y^i}{p^i} \frac{\partial p^i}{\partial \Theta_j} - \frac{1 - y^i}{1 - p^i} \frac{\partial p^i}{\partial \Theta_j} \right)
$$

–ü–æ–¥—Å—Ç–∞–≤–ª—è–µ–º $\frac{\partial p^i}{\partial \Theta_j} = p^i (1 - p^i) X_j^i$:

$$
= \sum_{i=1}^m \left( y^i (1 - p^i) - (1 - y^i) p^i \right) X_j^i
$$

 –£–ø—Ä–æ—â–∞–µ–º:

$$
= \sum_{i=1}^m \left( y^i - p^i \right) X_j^i
$$

–ï—Å–ª–∏ z>0 =>  ùë¶ÃÇ >0.5 => –≤—ã–±–∏—Ä–∞–µ–º 1 –∫–∞–∫ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ, –∏–Ω–∞—á–µ 0

### 4 Model evaluation

|   | y = 1 | y = 0 |
| ------------- | ------------- | ------------- |
| $\hat{y}=1$  | TP  | FP  |
| $\hat{y}=0$  | FN  | TN  |  

precision(—Ç–æ—á–Ω–æ—Å—Ç—å) = $\frac{TP}{TP+FP}$  
recal(–ø–æ–ª–Ω–æ—Ç–∞) = $\frac{TP}{TP+FN}$  


TPR(true positive rate) = $\frac{TP}{TP+FN}$ (x-axis - –æ—Å—å –∞–±—Å—Ü–∏—Å—Å x)  
FPR(false positive rate) = $\frac{FP}{FP+TN}$ (y-axis - –æ—Å—å –æ—Ä–¥–∏–Ω–∞—Ç x)  


1. –°–æ—Ä—Ç–∏—Ä—É–µ–º –æ–±—ä–µ–∫—Ç—ã –ø–æ —É–±—ã–≤–∞–Ω–∏—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∞ "1".  
2. –ù–∞—á–∏–Ω–∞–µ–º —Å —Ç–æ—á–∫–∏ (0, 0).  
3. –î–ª—è –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞:  
   - –ï—Å–ª–∏ –∏—Å—Ç–∏–Ω–Ω—ã–π –∫–ª–∞—Å—Å = 1 ‚Üí —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º TPR (–¥–≤–∏–≥–∞–µ–º—Å—è –≤–≤–µ—Ä—Ö).  
   - –ï—Å–ª–∏ –∏—Å—Ç–∏–Ω–Ω—ã–π –∫–ª–∞—Å—Å = 0 ‚Üí —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º FPR (–¥–≤–∏–≥–∞–µ–º—Å—è –≤–ø—Ä–∞–≤–æ).  
4. AUC-ROC ‚Äî –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –ø–æ–ª—É—á–∏–≤—à–µ–π—Å—è –∫—Ä–∏–≤–æ–π.  


$F_1 score$

$$
F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}} = \frac{2TP}{2TP + FP + FN}
$$

- **F‚ÇÅ** ‚Äî —ç—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ *–≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ* –º–µ–∂–¥—É Precision –∏ Recall.  
- –î–∞—ë—Ç –±–æ–ª—å—à–∏–π –≤–µ—Å –Ω–∏–∑–∫–∏–º –∑–Ω–∞—á–µ–Ω–∏—è–º: –µ—Å–ª–∏ Precision –∏–ª–∏ Recall –±–ª–∏–∑–æ–∫ –∫ 0, F‚ÇÅ —Ç–æ–∂–µ –±—É–¥–µ—Ç –±–ª–∏–∑–æ–∫ –∫ 0.  
- –ò–¥–µ–∞–ª—å–Ω–æ: **F‚ÇÅ = 1** (–∫–æ–≥–¥–∞ Precision –∏ Recall = 1).  
- –•—É–¥—à–∏–π —Å–ª—É—á–∞–π: **F‚ÇÅ = 0** (–µ—Å–ª–∏ TP = 0).

$F_\beta score$  

$$
F_\beta = (1 + \beta^2) \cdot \frac{\text{Precision} \cdot \text{Recall}}{(\beta^2 \cdot \text{Precision}) + \text{Recall}}
$$

- **Œ≤** ‚Äî –ø–∞—Ä–∞–º–µ—Ç—Ä, –æ–ø—Ä–µ–¥–µ–ª—è—é—â–∏–π –≤–∞–∂–Ω–æ—Å—Ç—å Recall –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ Precision:
  - **Œ≤ > 1**: –ë–æ–ª—å—à–∏–π –≤–µ—Å Recall (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–∞–∂–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å FN).  
  - **Œ≤ < 1**: –ë–æ–ª—å—à–∏–π –≤–µ—Å Precision (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤–∞–∂–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å FP).  
  - **Œ≤ = 1**: –≠–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ F‚ÇÅ-score.  
- –ü—Ä–∏–º–µ—Ä—ã:
  - **F‚ÇÇ-score** (Œ≤=2): Recall –≤ 2 —Ä–∞–∑–∞ –≤–∞–∂–Ω–µ–µ Precision.  
  - **F‚ÇÄ.‚ÇÖ-score** (Œ≤=0.5): Precision –≤ 2 —Ä–∞–∑–∞ –≤–∞–∂–Ω–µ–µ Recall.

```python
data = data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# –ó–∞–º–µ–Ω–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ Age –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º
data['Age'] = data['Age'].fillna(data['Age'].median())

# –ó–∞–º–µ–Ω–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ Embarked –º–æ–¥–æ–π
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

# –°–µ–º–µ–π–Ω—ã–π —Ä–∞–∑–º–µ—Ä
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1

# –û–¥–∏–Ω–æ—á–∫–∞ (1, –µ—Å–ª–∏ FamilySize = 1)
data['IsAlone'] = (data['FamilySize'] == 1).astype(int)

# Encoder
encoder = TargetEncoder(cols=['Sex', 'Embarked', 'FamilySize'])

data[['Sex', 'Embarked', 'FamilySize']] = encoder.fit_transform(
    data[['Sex', 'Embarked', 'FamilySize']], 
    data['Survived']  # –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è
)
```

### –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ

### L2-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è (Ridge)

–§—É–Ω–∫—Ü–∏—è —à—Ç—Ä–∞—Ñ–∞:
$$ R_{L2}(W) = \frac{\lambda_2}{2} \sum_{j=1}^{n} W_j^2 $$

–ì—Ä–∞–¥–∏–µ–Ω—Ç —à—Ç—Ä–∞—Ñ–∞:
$$ \nabla_W R_{L2} = \lambda_2 \cdot W $$

–û–±–Ω–æ–≤–ª–µ–Ω–∏–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞:
$$ \text{gradients} \leftarrow \text{gradients} - \frac{\lambda_2}{m} W $$

### L1-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è (Lasso)

–§—É–Ω–∫—Ü–∏—è —à—Ç—Ä–∞—Ñ–∞:
$$ R_{L1}(W) = \lambda_1 \sum_{j=1}^{n} |W_j| $$

–ì—Ä–∞–¥–∏–µ–Ω—Ç —à—Ç—Ä–∞—Ñ–∞:
$$ \nabla_W R_{L1} = \lambda_1 \cdot \text{sign}(W) $$

–≥–¥–µ —Ñ—É–Ω–∫—Ü–∏—è sign –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –∫–∞–∫:

$$
\text{sign}(W_j) =
\begin{cases} 
1, & \text{–µ—Å–ª–∏ } W_j > 0 \\ 
-1, & \text{–µ—Å–ª–∏ } W_j < 0 \\ 
0, & \text{–µ—Å–ª–∏ } W_j = 0 
\end{cases}
$$

–û–±–Ω–æ–≤–ª–µ–Ω–∏–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞:
$$ \text{gradients} \leftarrow \text{gradients} - \frac{\lambda_1}{m} \text{sign}(W) $$

$n$ ‚Äî —ç—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (–≤–µ—Å–æ–≤) –≤ –≤–µ–∫—Ç–æ—Ä–µ $W$. –¢–æ –µ—Å—Ç—å, –µ—Å–ª–∏ —É –Ω–∞—Å –µ—Å—Ç—å $W = (W_1, W_2, ..., W_n)$, —Ç–æ $n$ ‚Äî —ç—Ç–æ –æ–±—â–µ–µ —á–∏—Å–ª–æ —Ç–∞–∫–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

$m$ ‚Äî —ç—Ç–æ —Ä–∞–∑–º–µ—Ä –º–∏–Ω–∏-–±–∞—Ç—á–∞ (batch size), —Ç–æ –µ—Å—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–º–µ—Ä–æ–≤, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –≥—Ä–∞–¥–∏–µ–Ω—Ç –∑–∞ –æ–¥–∏–Ω —à–∞–≥ –æ–±—É—á–µ–Ω–∏—è

```python
class My_LogisticRegression:
    def __init__(self, step=0.01):
        self.W = None
        self.step = step

    def sigmoid(self, X):
        return 1.0 / (1.0 + np.exp(-X))  # –ò—Å–ø–æ–ª—å–∑—É–µ–º numpy.exp –¥–ª—è –≤–µ–∫—Ç–æ—Ä–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ü–∏–π

    def loss_function(self, X, W, y): # –º–∏–Ω—É—Å –ª–æ–≥–∞—Ä–∏—Ñ–º —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è

        y_pred = self.sigmoid(X @ W)
        y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)
        loss = -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
        return loss

    def fit(self, X, y, max_epochs=1000, tol=1e-6, reg="No", lambda_1=0, lambda_2=0, batch_size=1):
 
        X = np.hstack((np.ones((X.shape[0], 1)), X))  # –î–æ–±–∞–≤–ª—è–µ–º bias
        y = np.array(y)
        self.W = np.zeros(X.shape[1])  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –≤–µ—Å–æ–≤

        n_samples = X.shape[0]
        for epoch in range(max_epochs): # –ü—Ä–æ–±–µ–≥–∞–µ–º –ø–æ —ç–ø–æ—Ö–∞–º
            indices = np.arange(n_samples)
            np.random.shuffle(indices)
            
            for i in range(0, n_samples, batch_size):
                
                # –§–æ—Ä–º–∏—Ä—É–µ–º –º–∏–Ω–∏-–±–∞—Ç—á
                batch_indices = indices[i:i + batch_size]
                X_batch = X[batch_indices]
                y_batch = y[batch_indices]
                
                # –ü—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∏ –≤—ã—á–∏—Å–ª—è–µ–º –æ—à–∏–±–∫–∏
                y_pred = self.sigmoid(X_batch @ self.W)
                errors = y_batch - y_pred
                
                # –í—ã—á–∏—Å–ª—è–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç—ã
                gradients = X_batch.T @ errors / batch_size
                
                # –î–æ–±–∞–≤–ª—è–µ–º —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—é
                if reg == "l2":
                    gradients -= lambda_2 * self.W / batch_size
                elif reg == "l1":
                    gradients -= lambda_1 * np.sign(self.W) / batch_size

                # –û–±–Ω–æ–≤–ª—è–µ–º –≤–µ—Å–∞
                self.W += self.step * gradients

            # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—Ö–æ–¥–∏–º–æ—Å—Ç—å –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å
            loss = self.loss_function(X, self.W, y)
            if loss < tol:
                print(f"Converged at epoch {epoch}")
                break

    def predict_proba(self, X):

        X = np.hstack((np.ones((X.shape[0], 1)), X))  # –î–æ–±–∞–≤–ª—è–µ–º bias
        return self.sigmoid(X @ self.W)

    def predict(self, X):

        y_proba = self.predict_proba(X)
        return (y_proba >= 0.5).astype(int)
```

```python
%%time
# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
my_lr_model = My_LogisticRegression(step=0.1)
my_lr_model.fit(data.iloc[:, 1:].values, data['Survived'].values, max_epochs=100, batch_size=500)
```
