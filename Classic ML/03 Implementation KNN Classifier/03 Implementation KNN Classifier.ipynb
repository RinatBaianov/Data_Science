### 1 Definition

KNN-классификатор (k-nearest neighbors, метод k-ближайших соседей) — это алгоритм машинного обучения, который относит объект к тому классу, который наиболее распространён среди его k ближайших соседей в признаковом пространстве.

### 2 Basic algorithm

1) Сначала упорядочиваем данные по дереву (k-dimensional tree)   
2) Далее уже каждой точке находим n ближайших соседей (спускаемся и находим ближайшего соседа, далее поднимаемся и идём в другие соседние листы, чтобы убедиться чтотам расстояние не ближе)
3) Предсказание новой точки будет равно наиболее популярному таргету среди n соседей  


### 3 KD-Tree (k-dimensional tree) Construction Phase
Этап построения KDTree (k-dimensional tree)

KDTree (k-dimensional tree) — это структура данных, которая используется для быстрой обработки задач поиска ближайших соседей (k-Nearest Neighbors, kNN), вычисления расстояний и других операций, связанных с многомерными данными. Она особенно эффективна при работе с большими наборами данных в пространствах с низкой или средней размерностью.

Быстрота: Для низкоразмерных данных KDTree ускоряет поиск ближайших соседей. Вместо полного перебора всех точек (O(n)), поиск занимает в среднем O(log(n)) времени.
Гибкость: Поддерживает разные метрики расстояния (например, Евклидова, Манхэттенская, Минковского и т.д.).

### Алгоритм построения KD-Tree (k-dimensional tree)

#### Пример для 2D (k=2)

#### Основная идея

depth — глубина узла (начинается с 0 для корня).

k — количество измерений (осей) данных (например, для 2D точек k=2).

depth % k — остаток от деления глубины на k. Это определяет, по какой оси мы разделяем пространство в текущем узле

#### Итоговое дерево

Даны точки: [(2,3), (5,4), (9,6), (4,7), (8,1), (7,2)].

Глубина 0 (ось X):

Сортируем по X: [(2,3), (4,7), (5,4), (7,2), (8,1), (9,6)].

Определяем корень дерева: 
Медиана: (7,2) → корень дерева.

Левое поддерево (X < 7): [(2,3), (4,7), (5,4)].

Глубина 1 (ось Y): медиана (5,4).

Правое поддерево (X ≥ 7): [(8,1), (9,6)].

Глубина 1 (ось Y): медиана (9,6). 

Да, если расстояния между точками по первой координате совпадают (например, при сравнении точек), то алгоритм переходит к следующей координате для разрешения "ничьи"

### 4 Nearest neighbor search phase 
Этап поиска поиска ближайшего соседа

разберём поиск 3 ближайших соседей для точки **Q(4,5)** в KD-Tree с метрикой L1 (Манхэттенское расстояние), подсвечивая проверяемые узлы и объясняя логику.


#### Шаг 1: Спуск до листа
1. **Корень (7,2)** (ось X):
   - Сравниваем Q_x = 4 с 7 → 4 < 7 → идём влево
   - *Текущий путь*: [(7,2)]

2. **Узел (5,4)** (ось Y):
   - Сравниваем Q_y = 5 с 4 → 5 > 4 → идём вправо
   - *Текущий путь*: [(7,2), (5,4)]

3. **Лист (4,7)**:
   - Добавляем в кандидаты: (4,7)
   - Расстояние: L₁ = 2
   - *Текущий путь*: [(7,2), (5,4), (4,7)]
   
**Итог:** Нашли отправную точку (4,7), относительно которой будем искать ближайших соседей, теоретически она может не войти в топ-n ближайших соседей

#### Шаг 2: Обратный подъём
1. **Возврат к (5,4)**:
   - Расстояние до (5,4): |4-5| + |5-4| = 2
   - Обновляем кандидатов: [(4,7), (5,4)] (оба с расстоянием 2)
   - Проверяем левое поддерево (2,3):
     - Расстояние: |4-2| + |5-3| = 4
     - Обновляем кандидатов: [(4,7), (5,4), (2,3)]

2. **Возврат к корню (7,2)**:
   - Расстояние до (7,2): |4-7| + |5-2| = 6 (не в топ-3)
   - Проверяем возможность ближайших точек в правом поддереве:
     - Минимальное расстояние до правого подпространства (X≥7): |4-7| = 3
     - Текущее худшее расстояние в топ-3: 4
     - Поскольку 3 < 4, исследуем правую ветвь

3. **Правая ветвь**:
   - Узел (9,6):
     - Расстояние: |4-9| + |5-6| = 6 (не в топ-3)
   - Узел (8,1):
     - Расстояние: |4-8| + |5-1| = 8 (не в топ-3)

#### Итоговые 3 ближайших соседа:
1. (4,7) - расстояние 2
2. (5,4) - расстояние 2 
3. (2,3) - расстояние 4

#### Критерий останова

Здесь алгоритм может остановитья, просто в моем примере мы смотрим на разницу первых координат моей точки Q(4,5) и, например, узла (7,2) и получаем **|4-7| = 3**, а это меньше чем 4 (это самая дальная точка в нашем пуле по Манхеттеновскому расстоянию), если бы наш узел был больше (не 3, а 5), то мы бы не шли дальше и остановили поиск ближайших соседей

### Minkowski distance (or $L_p$-norm)  

Метрика Минковского (\( L_p \)-норма) — это обобщение таких известных метрик расстояния, как евклидова и манхэттенская. Она задаётся формулой:

$$
d(\mathbf{x}, \mathbf{y}) = \left( \sum_{i=1}^n |x_i - y_i|^p \right)^{1/p}
$$

#### Где:
- $\mathbf{x} = (x_1, x_2, \dots, x_n))$ и $\mathbf{y} = (y_1, y_2, \dots, y_n)$ — это два $n$-мерных вектора (например, точки в пространстве признаков);
- \( |x_i - y_i| \) — модуль разницы между соответствующими компонентами двух векторов;
- \( p \) — степень нормы, задающая форму расстояния.

---

#### Частные случаи метрики Минковского:
1. **\( p = 1 \)**: **Манхэттенское расстояние** (L1-норма):
   $
   d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^n |x_i - y_i|
   $
   Это сумма модулей разностей между координатами. Представляет "квадратное" расстояние по сетке.

2. **\( p = 2 \)**: **Евклидово расстояние** (L2-норма):
   $
   d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}
   $
   Это классическое расстояние в пространстве, измеряющее прямую линию между двумя точками.

3. **$p \to \infty $**: **Чебышёвское расстояние** (максимальная разница):
   $
   d(\mathbf{x}, \mathbf{y}) = \max_i |x_i - y_i|
   $
   Это максимальная разность между соответствующими компонентами двух векторов.

---

#### Свойства метрики Минковского:
- **Обобщаемость**: Она обобщает различные метрики (манхэттенскую, евклидову, чебышёвскую и другие) в зависимости от значения $p$.
- **Монотонность**: При увеличении $p$ значение расстояния возрастает или остаётся постоянным (влияет на приоритет крупных разностей).
- **Симметрия**: $d(\mathbf{x}, \mathbf{y}) = d(\mathbf{y}, \mathbf{x})$.
- **Неотрицательность**: $d(\mathbf{x}, \mathbf{y}) \geq 0$, причём $d(\mathbf{x}, \mathbf{y}) = 0 )$, если  $\mathbf{x} = \mathbf{y}$.
- **Неравенство треугольника**: $d(\mathbf{x}, \mathbf{z}) \leq d(\mathbf{x}, \mathbf{y}) + d(\mathbf{y}, \mathbf{z})$.

---

#### Примечание:
Метрика Минковского широко применяется в алгоритме \( k \)-ближайших соседей (KNN), где расстояние между точками является ключевым фактором для классификации или регрессии.

### 5 Implementation

```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score, roc_curve

from category_encoders import TargetEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from scipy.spatial import KDTree
```

| №  | Название колонки | Тип данных       | Описание |
|----|------------------|------------------|----------|
| 1  | `PassengerId`    | `int`            | Уникальный ID пассажира |
| 2  | `Survived`       | `int` (0/1)      | Выжил (1) или нет (0) — **таргет для предсказания** |
| 3  | `Pclass`         | `int` (1-3)      | Класс билета:<br>• 1 = 1-й класс<br>• 2 = 2-й класс<br>• 3 = 3-й класс |
| 4  | `Name`           | `string`         | Имя пассажира (включая титул, например, "Mr.", "Miss") |

| 6  | `Age`            | `float`          | Возраст (есть пропуски — `NaN`) |

```python
data = pd.read_csv(r"../00 Data/titanic.csv")

print(data.shape)
data.head()
```

```python
data = data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# Замена пропусков в Age медианным значением
data['Age'] = data['Age'].fillna(data['Age'].median())

# Замена пропусков в Embarked модой
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

# Семейный размер
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1

# Одиночка (1, если FamilySize = 1)
data['IsAlone'] = (data['FamilySize'] == 1).astype(int)

# Encoder
encoder = TargetEncoder(cols=['Sex', 'Embarked', 'FamilySize'])

data[['Sex', 'Embarked', 'FamilySize']] = encoder.fit_transform(
    data[['Sex', 'Embarked', 'FamilySize']], 
    data['Survived']  # Целевая переменная
)
```

```python
# StandardScaler для всех признаков
features = data.columns.drop('Survived')
data[features] = StandardScaler().fit_transform(data[features])
```

```python
data.head()
```

```python
class My_KNN:
    def __init__(self, n_neighbors=5, leaf_size=30):
        self.n_neighbors = n_neighbors
        self.tree = None
        self.X = None
        self.y = None
        self.leaf_size = leaf_size  # Учитываем параметр leaf_size
        
    def fit(self, X, y):

        self.X = X
        self.y = y
        self.tree = KDTree(X, leafsize=self.leaf_size)  # Передаем leaf_size в KDTree
        return

    def predict_proba(self, X2):

        distances, indices = self.tree.query(X2, k=self.n_neighbors)
        
        # Для каждого набора соседей вычисляем процентное соотношение класса 1
        neighbors_y = self.y[indices]  # Извлекаем классы соседей
        probas = np.mean(neighbors_y, axis=1)  # Для каждого теста вычисляем среднее (вероятность класса 1)
        
        return probas

    def predict(self, X):
        
        y_proba = self.predict_proba(X)  # Получаем вероятности
        return (y_proba >= 0.5).astype(int)  # Если вероятность >= 0.5, то класс 1, иначе 0
```

```python
%%time
# Инициализация и обучение модели
my_knn_model = My_KNN()
my_knn_model.fit(data.iloc[:, 1:].values, data['Survived'].values)
```

```python
%%time
my_y_proba = my_knn_model.predict_proba(data.iloc[:, 1:].values)
```

```python
%%time
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(data.iloc[:, 1:].values, data['Survived'].values)
```

```python
%%time
y_proba = knn_model.predict_proba(data.iloc[:, 1:].values)[:,1]
```

```python
# Вычисление AUC-ROC
AUC_my = roc_auc_score(data['Survived'].values, my_y_proba)
AUC_sclearn = roc_auc_score(data['Survived'].values, y_proba)

print(f"AUC-ROC (Custom Model): {AUC_my:.4f}")
print(f"AUC-ROC (Custom Model): {AUC_sclearn:.4f}")
```
