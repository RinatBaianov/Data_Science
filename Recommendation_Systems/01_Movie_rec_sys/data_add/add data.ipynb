### 4 Обработка данных для того, чтобы оцифрованные данные использовались в модели

#### 4.1 Топ 20 Directors

```python
def encode_top_N_column(df, column, N=20):
    """
    Преобразует столбец с режиссёрами/актёрами/и т.д. в one-hot матрицу
    по топ-N самым популярным значениями.

    Аргументы:
        df (pd.DataFrame): таблица с данными
        column (str): имя столбца с перечисленными значениями (строка с запятыми)
        N (int): сколько самых популярных значений учитывать (по умолчанию 20)

    Возвращает:
        pd.DataFrame: one-hot DataFrame с колонками top-N значений
    """
    # Разбиваем строки по запятой, убираем пробелы, разворачиваем в один список
    all_items = df[column].dropna().str.split(',').explode().str.strip()
    
    # Находим топ-N самых популярных
    top_N_items = all_items.value_counts().head(N).index.tolist()

    # Функция, возвращающая строку one-hot по топ-N
    def encode_row(cell):
        row = [0] * len(top_N_items)
        if pd.isna(cell):
            return row
        for i, item in enumerate(top_N_items):
            if item in cell:
                row[i] = 1
        return row

    # Применяем функцию ко всему столбцу
    encoded = df[column].apply(encode_row)
    encoded_df = pd.DataFrame(encoded.tolist(), columns=[f"{column}_{name}" for name in top_N_items])

    return encoded_df
```

```python
useful_df_Director = encode_top_N_column(useful_df, "Director", N=20)

print(useful_df_Director.shape)
useful_df_Director.head()
```

```python
useful_df = pd.concat([useful_df, useful_df_Director], axis=1)
print(useful_df.shape)
useful_df.head()
```

#### 4.4 Посмотрим на те столбцы у кого есть пустые значения и какой % непустых значений в этих столбцах

```python
null_stats = useful_df.isnull().mean()  # Доля пропусков по каждому столбцу
cols_with_nulls = null_stats[null_stats > 0]  # Только столбцы с пропусками
percent_non_nulls = (1 - cols_with_nulls) * 100  # Процент непустых

# Вывод
for col, percent in percent_non_nulls.items():
    print(f"{col}: {percent:.2f}% непустых значений")
```
